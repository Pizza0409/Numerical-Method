```python
#define step size
h = 1
#define number of iteration
iterations = 20
#list of the step sizes
step_size = []
#list of the max error for each steps
max_error = []
```

```python
import numpy as np
# implement forward difference
# y = f(x)
# y' = [f(x_i+1)- f(x_i)]/h
for i in range(iterations): 
    #append the step size to list step_size
    step_size.append(h)
    #decreasing the step size
    h /= 2 # h = h/2
    #the computing grids
    x = np.arange(0, 2*np.pi, h)
    #compute the value of function cos(x)
    y = np.cos(x)
    #compute the differences
    forward_diff = np.diff(y)/h
    #compute x
    x_diff = x[:-1]
    #exact solution
    exact_solution = -np.sin(x_diff)
    
    
    #compute the max error
    error_max= max(abs(exact_solution - forward_diff))
    #append the max error
    max_error.append(error_max)
    
    
```

```python
#plot log-log plot
import matplotlib.pyplot as plt
plt.style.use("seaborn-poster")
max_error
step_size
plt.loglog(step_size,max_error,"v")
```
