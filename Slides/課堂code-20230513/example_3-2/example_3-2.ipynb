```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
plt.style.use("seaborn-poster")
```

```python
#given theta is on [0,2pi]
#set forward step
dx = 0.1
#deccritized your grid
x = np.arange(0,2*np.pi,dx)
y = np.cos(x)
```

```python
#compute forward difference
forward_diff = np.diff(y)/dx
x_diff = x[:-1:]
```

```python
plt.figure
plt.plot(x_diff,forward_diff, "-", label = "finite difference approximation")
```

```python
#compute exact solution
exact = -np.sin(x_diff)
plt.plot(x_diff,exact,label = "exact solution")
```

```python
plt.figure
plt.plot(x_diff,forward_diff, "-.", label = "finite difference approximation")
plt.plot(x_diff,exact,label = "exact solution")
plt.legend()
```

```python
#compute error
error = abs(exact-forward_diff)
```

```python
plt.plot(x_diff,error)
```
