```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def central_diff(y):
    # y : compute function
    # h : step size
    h = 0.1
    n = len(y)
    d = np.zeros(n)                          #creat a zero matrix which the len is y
    
    for i in range(1, n - 1):              
        d[i] = (y[i+1] - y[i-1]) / (2 * h)   # f'(xi) = (f(xj+1) - f(xj-1))/2h

    return d
```

```python
def Central_Difference_Smoothing(x, y, n):
    y_smooth = np.zeros(len(y))    
        
    d = np.zeros(len(y))
    h = 0.1
    
    #y_smooth
    for i in range(1, len(y) - 1):
        y_smooth[i-n] = np.mean(y[i-n:i+n])
    
    #central diff
    d = central_diff(y_smooth)

    return d, x
```

```python
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x) + np.random.randn(len(x))/100
dy, X = Central_Difference_Smoothing(x, y, n=4)
```

```python
plt.plot(x, y)
plt.title("Noisy Sine Function")
plt.xlabel("x")
plt.ylabel("y")
plt.show
```

```python
z = np.cos(x)
w = central_diff(y)
plt.plot(X, dy, label = "smoothed")
plt.plot(x, z, label = "cosine")
plt.plot(x, w, label = "unsmoothed forward diff")
plt.title("Analytic Derivative and Smoothed Derivate")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show
```
